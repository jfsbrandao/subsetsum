#Algoritmo proposto

```python
conjuntodesor =  []
with open('G1-10','r') as arquivo:
  for valor in arquivo:
    valor = valor.rstrip('\n')
    conjuntodesor.append(int(valor))
```

```python
def subsetsum(array, num):
    result = []
    def find(arr, num, path=()):
        if not arr:
            return
        if arr[0] == num:
            result.append(path + (arr[0],))
        else:
            find(arr[1:], num - arr[0], path + (arr[0],))
            find(arr[1:], num, path)
    find(array, num)
    return result

array = [-3,5,12,-1,-4,2]
num = -7
print(subsetsum(array,num))
```

```python
def subsetsum(array, num):
    result = []
    def find(arr, num, path=()):
        if not arr:
            return
        if arr[0] == num:
            result.append(path + (arr[0],))
        else:
            find(arr[1:], num - arr[0], path + (arr[0],))
            find(arr[1:], num, path)
    find(array, num)
    return result

num = 19
print(subsetsum(conjuntodesor,num))
```

```python
import matplotlib.pyplot as plt
import numpy as np
def complexidade(w):
    return (2**w)*w

x = np.linspace(0, 10, 20)
y = complexidade(x)

plt.xlabel('Tamhano do conjunto')
plt.ylabel('Número de processos')
plt.plot(x, y)
plt.plot(x, y, 'o')
plt.show()
```
